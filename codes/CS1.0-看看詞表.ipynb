```python
filename = "common_words.txt"
with open(filename, 'r') as F:
  data = F.readlines()
```

```python
# 轉繁體
!pip install opencc-python-reimplemented
```

```python
from opencc import OpenCC
cc = OpenCC('s2t')
```

```python
cmdict = {}
for entry in data:
  entry = entry.split('\t')
  cmword = entry[0]
  cmword = cc.convert(cmword)
  wfreq = int(entry[-1])
  # discard pinyin 
  cmdict[cmword] = wfreq  
```
