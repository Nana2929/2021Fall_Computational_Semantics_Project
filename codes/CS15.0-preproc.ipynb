````python
class MSW:
    def __init__(self, word):
        self.word = word
        self.senses = [] # list of dicts 
        self.N = len(self.senses)
        self.id = -1
        self.haslevel = False
    def updateN(self):
        self.N = len(self.senses)
        return len(self.senses)
    def updateS(self, senses):
        '''a list of dicts'''
        self.senses = senses 
        self.updateN()  
    def updateID(self, id):
        self.id = id 
    def updateLevel(self, haslevel):
        self.haslevel = haslevel
    def __str__(self):
        return f"* word: {self.word},* #senses:{self.N} *hasLevel: {self.haslevel}"
    __repr__ = __str__
````

````python
# Wiki zh prefix: https://zh.wikipedia.org
MSW_list = []
Coarse_MSW = []
for idx, msw in enumerate(list(MSW_dict.keys())):
    # wi 
    word = msw 
    sense_list = MSW_dict[word]
    delmain = False
    leveled = False
    mainslink = ""
    org_senselist = []
    # print(word)
    for sense_entry in sense_list:
          # print(sense_entry)
          cat = sense_entry[0]
          # check if cat is not a real category
          if cat in ["參見", "另見", "相關條目"]:
              cat = ""
          if cat != "" and cat != "main":
              leveled = True
          sense = sense_entry[1]
          views = sense_entry[2]
          gloss = sense_entry[3]
          sublink = sense_entry[4]
          ismain = False
          # 另見 相關條目 
          # 然後有空白的category都放在 其他
          if cat == "main": # main一定是第一個
              mains_link = sublink
              ismain = True
          if sublink == mainslink:
              ismain = True
              delmain = True
          # check if link is valid 
          # if isValid(sublink):
          #   link = f"https://zh.wikipedia.org{sublink}"
          # else:
          #   link = "" 
          # some redirect link
          # construct dictionary for this sense (wi.sj)
          link = ""
          if sublink.startswith("/wiki/"):
            link = f"https://zh.wikipedia.org{sublink}"
          sensedict = {
              'category': cat,
              'sense': sense,
              'views': views,
              'gloss': gloss, 
              'link': link,
              'ismain': ismain
          }
          org_senselist.append(sensedict)
    if delmain:
        del org_senselist[0] 
    print(*org_senselist, sep = '\n')
    msw = MSW(word)
    msw.updateS(org_senselist)
    msw.updateID(idx)
    msw.updateLevel(leveled)
    MSW_list.append(msw)
    if leveled: 
        Coarse_MSW.append(msw)
    print(msw)
````

````python
import wikipediaapi
wiki_wiki = wikipediaapi.Wikipedia('zh')
import os, sys
from opencc import OpenCC
cc = OpenCC('s2t')
````

````python
apple = wiki_wiki.page("蘋果")
def print_sections(sections, level=0):
    for s in sections:
         print("%s: %s - %s" % ("*" * (level + 1), s.title, s.text[0:40]))
         print_sections(s.sections, level + 1)
print_sections(apple.sections)
````
