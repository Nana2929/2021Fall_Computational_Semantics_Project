```python
wiki_wiki = wikipediaapi.Wikipedia('zh')
apple = wiki_wiki.page('蘋果') # 簡繁皆可欸 
print(f"Page - Exists: %s" % apple.exists())
print("Page - Title: %s" % apple.title)
print("Page - Summary: %s" % apple.summary[0:60]) # 好爽...謝謝wikipediaapi creditor 
```

```python
# apple_text = wiki_wiki.page("蘋果")
# apple_text.text
def print_links(page):
    links = page.links
    for title in links.keys():
        print("%s: %s" % (title, links[title]))
        if title == "苹果 (消歧义)":
          print("HIT!")
          apple_disambig = links[title]
          print(apple_disambig)
          return apple_disambig
apple_disambig = print_links(apple)
# print_links(apple_text) # 有消歧義的就可以用
# It’s map, where key is page title and value is WikipediaPage.
```

```python
print(f"Page - Exists: %s" % apple_disambig.exists())
print("Page - Title: %s" % apple_disambig.title)
print("Page - Summary: %s" % apple_disambig.summary[0:60])
print_links(apple_disambig)
```
